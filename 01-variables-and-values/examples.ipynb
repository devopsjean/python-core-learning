# Variables and Values

Observe assignment, comparison, and reassignment step by step.

## Step 1
Start by assigning three values to three variables. This is the first moment where names and values become connected.

```python
course_name = "python core"
study_minutes = 45
is_active = True

print("Assignment complete")
```

## Step 2
Check the current value and type before changing anything. This helps separate the idea of a variable name from the data it currently holds.

```python
print("Before reassignment")
print("course_name:", course_name, "| type:", type(course_name))
print("study_minutes:", study_minutes, "| type:", type(study_minutes))
print("is_active:", is_active, "| type:", type(is_active))
```

## Step 3
Self-check: use `==` to compare values without changing them. The result of comparison is always `True` or `False`.

```python
print("Comparison")
print("course_name == 'python core':", course_name == "python core")
print("study_minutes == 60:", study_minutes == 60)
print("is_active == True:", is_active == True)
```

## Step 4
Reassign two variables and inspect them again. Notice that the variable name stays the same, but the current value changes.

```python
study_minutes = study_minutes + 15
course_name = "python review"

print("After reassignment")
print("course_name:", course_name, "| type:", type(course_name))
print("study_minutes:", study_minutes, "| type:", type(study_minutes))
print("is_active:", is_active, "| type:", type(is_active))
print("study_minutes == 60:", study_minutes == 60)
```
