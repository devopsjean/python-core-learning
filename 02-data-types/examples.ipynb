# Data Types

Check how Python treats numbers, text, and booleans.

## Step 1
Create values with different data types and inspect them. Focus on what kind of value each variable holds.

```python
item_count = 3
price = 19.99
topic = "functions"
is_ready = False

print(type(item_count))
print(type(price))
print(type(topic))
print(type(is_ready))
```

## Step 2
Convert text into a number. This is useful when a value looks numeric but is still stored as a string.

```python
raw_number = "42"
converted_number = int(raw_number)
total = converted_number + item_count

print("raw_number:", raw_number, "| type:", type(raw_number))
print("converted_number:", converted_number, "| type:", type(converted_number))
print("total:", total)
```

## Step 3
Compare string combination with numeric addition. The output shows why type matters in Python.

```python
print(topic + " practice")
print("10" + "5")
print(10 + 5)
```

## Step 4
Convert values in the other direction as well. This shows that Python can turn numbers into text and text into floating-point numbers.

```python
number_as_text = str(15)
float_value = float("3.14")

print("number_as_text:", number_as_text, "| type:", type(number_as_text))
print("float_value:", float_value, "| type:", type(float_value))
```
